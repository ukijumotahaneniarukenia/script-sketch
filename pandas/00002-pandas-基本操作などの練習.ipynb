```python
df['year'] = list(pd.Series(df.index).apply(lambda x: x.year))
```

```python
df['month'] = list(pd.Series(df.index).apply(lambda x: x.month))
```

```python
df['day'] = list(pd.Series(df.index).apply(lambda x: x.day))
```

```python
df.head()
```

```python
help(pd.cut)
```

```python
labels = ['上旬', '中旬', '下旬']
df['period'] = pd.cut(list(df['day']),  bins=[0,10,20,31], labels=labels, right=True) # 0<day≦10, 10<day≦20, 20<day≦31
```

```python
df.head()
```

```python
df["precipitation"] = df["precipitation"].apply(lambda x : -1 if x == "--" else float(x))
```

```python
pd.DataFrame(df["precipitation"].unique(),columns=["precipitation"])
```

```python
defaultNanKeyword="none"
df["payday"]=df["payday"].fillna(0.0)
df["event"] = df["event"].fillna(defaultNanKeyword)
df["kcal"] = df["kcal"].fillna(df["kcal"].mean())
df["remarks"] = df["remarks"].fillna(defaultNanKeyword)
```

```python
df.isnull().sum()
```

```python
help(pd.DataFrame.diff)
```

```python
df['temperature_diff'] = df['temperature'].diff(periods=1)
```

```python
df[['temperature','temperature_diff']].head()
```

```python
help(pd.DataFrame.rolling)
```

```python
df['temperature_rolling_mean'] = df['temperature'].rolling(window=3).mean()
df[['temperature', 'temperature_diff', 'temperature_rolling_mean']].head()
```
