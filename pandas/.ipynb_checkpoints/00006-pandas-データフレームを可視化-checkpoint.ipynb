```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
help(pd.DataFrame.plot)
```

```python
df = pd.read_csv('iris.csv')
```

```python
df
```

```python
#ファイル保存　https://note.nkmk.me/python-pandas-plot/
plt.figure()
df.plot()
plt.savefig('01-iris-シングルプロット.png')
plt.close('all')
```

```python
#サブプロットの描画
plt.figure()
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(9, 6))
df.plot(ax=axes[0, 0], legend=False)
df.plot(ax=axes[1, 2], legend=False, kind='hist')
fig.savefig('02-iris-マルチプロット.png')
plt.close('all')
```
