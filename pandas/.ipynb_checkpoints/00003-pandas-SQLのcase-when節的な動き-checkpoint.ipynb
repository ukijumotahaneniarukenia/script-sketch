```python
# データの扱いに必要なライブラリ
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv('sample.csv',encoding="utf-8")
```

```python
df
```

```python
def rankTemperature(x):
    if  x > 20.0:
        return "A"
    elif x <= 10.0 and x <=20.0:
        return "B"
    else:
        return "C"
```

```python
df['rankTemperature'] = df['temperature'].apply(rankTemperature)
```

```python
df[["rankTemperature","temperature"]].groupby(["rankTemperature"]).count()
```

```python
df
```

- 単一分岐はラムダのapplyでonelinerにする必要がある

```python
def flgTemperature(x):
    if  x > 20.0:
        return 1
    else:
        return 0
```

```python
df['flgTemperature'] = df['temperature'].apply(lambda x : 1 if x > 20 else 0)
```

```python
df[["flgTemperature","temperature"]].groupby(["flgTemperature"]).count()
```

```python
df
```

```python
df["weather"].unique()
```

```python
def maskWeather(x):
    if  x == "薄曇":
        return "曇り"
    elif x == "曇":
        return "曇り"
    else:
        return x
```

```python
df['maskWeather'] = df['weather'].apply(maskWeather)
```

```python
df[["maskWeather","weather"]].groupby(["maskWeather"]).count()
```

- 対象行の複数列を用いて評価演算したい場合は、仮引数を指定してaxis=1（列方向）を指定する必要がある

```python
def subRank(row):
    if  (row.temperature > 20.0) & (row.weather == "快晴") :
        return "SA"
    elif (row.temperature >= 10.0 and row.temperature <=20.0) & (row.weather == "快晴"):
        return "SB"
    else:
        return "C"
```

```python
df['subRank'] = df.apply(lambda row:subRank(row),axis=1)
```

```python
df[["temperature","weather","subRank"]].drop_duplicates()
```

```python
def subRank2(row,upperLimit,lowerLimit):
    if  (row.temperature > upperLimit) & (row.weather == "快晴") :
        return "SA"
    elif (row.temperature >= lowerLimit and row.temperature <=upperLimit) & (row.weather == "快晴"):
        return "SB"
    else:
        return "C"
```

```python
upperLimit=20.0
lowerLimit=10.0
```

```python
df['subRank2'] = df.apply(lambda row:subRank2(row,upperLimit,lowerLimit),axis=1)
```

```python
df[["temperature","weather","subRank","subRank2"]].drop_duplicates()
```

```python
df[df["subRank"]!=df["subRank2"]]
```
