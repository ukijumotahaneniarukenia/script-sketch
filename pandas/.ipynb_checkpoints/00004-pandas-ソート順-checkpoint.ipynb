```python
list(df["weather"].unique())
```

```python
sort_weather_order={'快晴':4, '曇':3, '晴れ':5, '薄曇':7, '雨':1, '雪':2, '雷電':6}
```

```python
sort_weather_order
```

```python
df["sort_weather_order"]=sorted(df["weather"], key=lambda x: sort_weather_order[x])
```

```python
df.sort_values(by=["week","sort_weather_order","temperature"], ascending=[True,True,False])[["week","sort_weather_order","temperature"]]
```
